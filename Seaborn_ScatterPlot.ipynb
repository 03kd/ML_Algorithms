```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

## Goal: which gender have more no. of surviver?

```python
# Lets plot the data
plt.figure(figsize =(10,7))
sns.scatterplot(x ='age', y='fare', data = titanic_df)
```

##### In these plot we can define which is male and female.

### 1. hue

```python
plt.rcParams['figure.figsize'] = [10,7]
```

```python
sns.scatterplot(x ='age', y='fare', data = titanic_df, hue ='sex')
```

##### we can divide the age into groups. so, we use 'who' column

### 2. style

```python
sns.scatterplot(x='age', y='fare',  data=titanic_df, hue ='sex', style ='who')
```

### 3. size;- to increase the size of male and female

```python
sns.scatterplot(x='age', y='fare',  data=titanic_df, hue ='sex', style ='who', size ='who')
```

### 4. sizes;- to increase the size of markers in plot

```python
sns.scatterplot(x='age', y='fare',  data=titanic_df, hue ='sex', style ='who', size='who', sizes=(50,100))
```

## Goal

```python
sns.scatterplot(x='who', y='fare', data= titanic_df, hue ='alive', style ='alive', size='who', sizes =(60,80))
```

**Intrepretation** ;- As above plot, we can see that more no. of males is not alive and also number of womens and children are saved. There is an interesting fact that is, which men pay high fare are saved and low fare paid by women is saved.

```python
sns.scatterplot(x='who', y='fare', data= titanic_df, hue ='alive', style ='alive', size='who', sizes =(60,80),
               palette ='inferno_r', alpha = 0.6)
```
