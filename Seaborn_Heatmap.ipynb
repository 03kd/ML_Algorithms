### Seaborn used in Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
```

```python
# showing the dataset in heatmap
global_warming = pd.read_csv("Who_is_responsible_for_global_warming.csv")
global_warming
```

```python
# To show, drop some columns
global_warming = global_warming.drop(columns =['Country Code', 'Indicator Name', 'Indicator Code']).set_index('Country Name')
global_warming.head()
```

```python
plt.figure(figsize =(16,9))
sns.heatmap(global_warming)
```

```python
# Set the min and max value of color bar
plt.figure(figsize =(16,9))
sns.heatmap(global_warming, vmin =0, vmax = 21)
```

```python
# Change the color of heatmap
plt.figure(figsize =(16,9))
# sns.heatmap(global_warming, cmap ='fg') # fg is used because we don't know the color code so,
sns.heatmap(global_warming, cmap ='coolwarm')
```

```python
# To show the value in heatmap
plt.figure(figsize =(16,9))
sns.heatmap(global_warming, cmap ='coolwarm', annot =True)
```
