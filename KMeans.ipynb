```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x = -2* np.random.rand(200,2)
```

```python
x0 = 1+2*np.random.rand(100,2)
x0
```

```python
x[100:200] =x0
```

```python
plt.scatter(x[:,0], x[:,1],s =25, color ='r')
plt.grid()
```

```python
Kmean = KMeans(n_clusters =2)
Kmean.fit(x)
```

```python
Kmean.cluster_centers_
```
